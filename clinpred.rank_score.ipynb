```python
import pandas as pd
```

```python
df = pd.read_table('ClinPred-基因变异致病性功能预测.txt')
```

```python
df.columns
```

```python
df.head()
```

```python
df['rank'] = df.ClinPred_Score.rank()
```

```python
df = df.sort_values(by=['rank'],ascending=False)
```

```python
dt = df['ClinPred_Score'].dropna().value_counts().to_dict()
```

```python
df['rank_count'] = df[['ClinPred_Score']].applymap(lambda x:dt[x])
```

```python
df.head()
```

```python
df = df.drop_duplicates(['ClinPred_Score'])
```

```python
s = df.rank_count.sum()
```

```python
df['cumsum'] = df.rank_count.cumsum()
```

```python
df['rank_score'] = df[['cumsum']].applymap(lambda x:1 - x/s).head()
```

```python
df.to_csv('clinpred.rank_score.xls', sep='\t', index=False)
```
