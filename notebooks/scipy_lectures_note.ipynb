### Fibonacci sequence with lite dynamic programming

def fibo(n):
    a = 0; b = 1
    for i in range(n):
        a, b = b, a + b
    return b
for i in range(10):
    print(fibo(i))

### `quick sort` python code

```python
array = list(np.random.randint(100, size=20))

def quick_sort(array):
    if len(array) < 2:
        return array
    pivot = array[len(array) - 1]
    less = []
    greater = []
    array.remove(pivot)
    for i in array:
        if i < pivot + 1:
            less.append(i)
        else:
            greater.append(i)
    print(less ,[pivot] , greater)
    return quick_sort(less) + [pivot] + quick_sort(greater)

print(quick_sort(array))
```
